## Check difference between models

```python
stats = {}

for MAG in GEMs_adapt.keys():
    #find number of reactions in each model
    rxns1 = GEMs_adapt_media[MAG].reactions
    rxns2 = GEMs_adapt_media2[MAG].reactions
    rxns3 = GEMs_adapt_media3[MAG].reactions
   
    
    #find intersection between models
    intersection = set.intersection(*map(set,[rxns1,rxns2,rxns3]))
    
    stats[(MAG,1)]=len(rxns1)
    stats[(MAG,2)]=len(rxns2)
    stats[(MAG,3)]=len(rxns3)

    stats[(MAG,"intersection")] = len(intersection)
    
    
```

```python
difference_stats = pd.Series(stats).unstack()

difference_stats["max_difference"] = difference_stats.apply(lambda row:max([row[1]-row["intersection"],row[2]-row["intersection"],row[3]-row["intersection"]]),
                                                       axis=1)
difference_stats.sort_values("max_difference",ascending=False).head(20)                                              
```

```python
difference_stats[difference_stats.max_difference>0].shape
```

```python
rxns_added = {}
count_GEMs_rxns_added = {}
for MAG in GEMs_adapt.keys():
    GEMs_adapt_rxns = set(GEMs_adapt[MAG].reactions.keys())
    GEMs_adapt_media_rxns = set(GEMs_adapt_media[MAG].reactions.keys())
    
    difference = GEMs_adapt_media_rxns-GEMs_adapt_rxns
    
    count_GEMs_rxns_added[MAG]=len(difference)
    print(MAG)
    for rxn in difference:
        
        if rxn in rxns_added.keys():
            rxns_added[rxn]=rxns_added[rxn]+1
        else:
            rxns_added[rxn]=1
        if rxn=="R_IPPS":
            print(GEMs_adapt_media[MAG].reactions[rxn])
    
```
