# Analyzing GEMs from MAGs

```python
import reframed
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances
```

```python
import os
```

## No constraints

```python
directory = os 

directory = os.fsencode("../output/GEMs/GEMs_no_constraints/")

GEMs_dict = {}
for file in os.listdir(directory):
    print(file)
    
    filename = os.fsdecode(file)
    if filename.endswith(".xml"): 
        print(filename)
        GEMs_dict[filename[:-4]]= reframed.load_cbmodel("../output/GEMs/GEMs_no_constraints/"+filename)
        continue
    else:
        continue
```

```python
GEMs_stats = {}

for model_id, model in GEMs_dict.items():
    print(model_id)
    GEMs_stats[model_id]={"reactions":len(model.reactions.keys()),
                          "metabolites":len(model.metabolites.keys()),
                         "genes": len(model.genes.keys())}
```

```python
pd.DataFrame(GEMs_stats).transpose().sort_values(by="genes").plot(kind="hist",bins=100)
```
