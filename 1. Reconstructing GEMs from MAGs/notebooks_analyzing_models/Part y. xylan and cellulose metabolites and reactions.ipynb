### Gapfilling for the Xylan based communities

```python
import pandas as pd


import reframed


import sys
sys.path.append("../functions/")
import general_functions as general_func

import colors_MAGs
```

```python
all_mags_paper = general_func.read_allmags_data()
```

```python
product = pd.read_csv("../input/files_from_fairdomhub/product.tsv",sep="\t",index_col="genome")
product
```

```python
xylan_degraders = product.loc[:,product.columns.str.contains("CAZy:")]
xylan_degraders
```

```python
MAG2community_id = pd.read_csv("../output/MAG2community_id.tsv",sep="\t",index_col=0,header=None)
```

```python
community_degraders = xylan_degraders.loc[MAG2community_id.groupby(1).groups["CD_X"]].sum(axis=1)

results={}
for MAG in MAG2community_id.groupby(1).groups["CD_X"]:
    print(MAG)
    
    model = reframed.load_cbmodel("../output/GEMs/GEMs_final_test/"+MAG+".xml")
    
    results[MAG]={"Family":all_mags_paper.loc[MAG,"Family"],
                  "Maria: polysaccharide degrades":community_degraders.loc[MAG],
                  "M_xylb_e":"M_xylb_e" in model.metabolites,
                  "M_xyl3_e":"M_xyl3_e" in model.metabolites,
                  "M_xyl4_e":"M_xyl4_e" in model.metabolites,
                  "M_xylan4_e":"M_xylan4_e" in model.metabolites,
                  "M_xylan8_e":"M_xylan8_e" in model.metabolites}
```

```python
pd.DataFrame(results).T
```

```python
pd.DataFrame(results).T.sort_values(by="Maria: polysaccharide degrades")
```
